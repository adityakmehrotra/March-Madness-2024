```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.common.exceptions import NoSuchElementException
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression, LogisticRegression
from bs4 import BeautifulSoup
import pandas as pd
import csv
import time
```

```python
all_rows = []

season = 2003
while (season <= 2023):
    with open('%d_meta_data_id.csv' % (season), 'r', newline = '') as file:
        reader = csv.reader(file)

        int_val = 0
        for row in reader:
            if (int_val != 0):
                all_rows.append(row)
            elif (season == 2003):
                all_rows.append(row)

                
            int_val += 1

    season += 1

print(all_rows)
        

with open('all_meta_data_id.csv', 'w', newline = '') as file:
        writer = csv.writer(file)

        for row in all_rows:
            writer.writerow(row)
```
